```python
import json
from pathlib import Path

from entities import Project

project_json_path = Path("../data/annotated/seven332|EhViewer.json")
```

```python
with open(project_json_path) as inf:
    project = Project(**json.load(inf))
```

```python
"""Export the dependency graph to json"""
def graph2json(graph):
    # Map the original node IDs to new numeric IDs
    id_mapping = {idx:graph.nodes[idx] for idx, node in enumerate(graph.nodes)}
    return id_mapping
    # Create the nodes dictionary using the numeric IDs and the original 'name' attributes
    nodes = {id_mapping[node]: node for node in graph.nodes}
    
    # Create the edges list using the numeric IDs for each node
    edges = [(id_mapping[u], id_mapping[v]) for u, v, _ in graph.edges]
    
    # Combine the nodes and edges into the desired dictionary format
    graph_json = {
        "nodes": nodes,
        "edges": edges
    }
    
    # Convert the dictionary to JSON format
    graph_json_str = json.dumps(graph_json)
    
    return graph_json_str
```

```python
json_str = graph2json(project.dep_graph)
```
